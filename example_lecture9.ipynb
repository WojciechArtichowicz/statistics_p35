```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
from random import randrange
%matplotlib inline
```

**Moivre-Laplace integral theorem**

Approximate empirical proof of the theorem

```python
p = 0.25
n = 1000000
BinomialDistribution = st.binom(n,p)
X = BinomialDistribution.rvs(n)
Y = (X - n*p)/(n*p*(1-p))**0.5

plt.hist(Y,bins=50,normed=True,edgecolor='black')
NormalDistribution = st.norm(0,1)
x = np.arange(-4,4,0.05)
plt.plot(x,NormalDistribution.pdf(x))
```

**Central limit theorem**

Approximate empirical proof of the theorem

```python
N= 10000
n = 10000 # sample size
U = np.zeros((N,)) #number of samples

SomeDistribution = st.expon(1/10) #use any distribution
m = SomeDistribution.mean()
s = SomeDistribution.std()

for i in range(N):
    X = SomeDistribution.rvs(n)
    U[i] = np.mean(X)

plt.hist(U,bins=50,normed=True,edgecolor='black')
S = s/n**0.5
NormalDistribution = st.norm(m,S)
x = np.arange(m-4*S,m+4*S,0.01*S)
plt.plot(x,NormalDistribution.pdf(x))
```

**Chebyshev theorem**

```python
ProbabilityDistribution = st.norm(0,1) # choose any distribution
#ProbabilityDistribution = st.expon()
m = ProbabilityDistribution.mean()
s = ProbabilityDistribution.std()

k = np.arange(1,10,0.1)
Pcheb = 1-1/k**2
Pexact = ProbabilityDistribution.cdf(m+k*s)- ProbabilityDistribution.cdf(m-k*s)
plt.plot(k,Pcheb,"red",k,Pexact,"blue")
plt.legend(["Chebyshev", "Exact"])
```

**Khinchin law of large numbers**

```python
#ProbabilityDistribution = st.norm(3,1) 
ProbabilityDistribution = st.expon() # choose any distribution

m = ProbabilityDistribution.mean()
N = 2000
Y = np.zeros(N)
for n in range(N):
    Y[n] = np.mean(ProbabilityDistribution.rvs(n+1))
    
plt.plot(np.array(range(1,N+1)),Y,"blue",[1,N],[m,m],"red")
plt.legend(["Y","$\mu$"])
```
